```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('Data for python.csv', sep=';', header=1, index_col=0, decimal=',')
#print all of df
df
```

```python
x = df["Id"]

y = df["Total Row Score C"]

```

```python
x1 = []
for i in range (123):
    if x[i]=='b':
        x1.append(1)
    else: x1.append(0)

x1 = np.array(x1).reshape(-1,1)
plt.plot(y,x1, 'o')
```

```python
model = LinearRegression().fit(x1, y)
model.summary()
```

```python
import statsmodels.api as sm

results = sm.OLS(y,x1).fit()
results.summary()  
```
