```python
from sklearn.model_selection import train_test_split
import numpy as np
from sklearn import metrics
from sklearn.preprocessing import MinMaxScaler
import pandas as pd
```

# Andre statiske modeller 

I dette dokument undersøger vi tre andre modeller som kan blive brugt til binær klassificering. Ved undersøgelse af disse modeller er konklusionen at resultaterne er lige så gode som fra logistisk regression. Grunden bag at vi valgte  logistisk regression er at den er hurtig og effektiv. 

```python
df = pd.read_csv('Behandlet data-kopi.csv', sep=';', header=0, index_col=0, decimal=',')
df
```

```python
def pin_update(row):
    if row["Total Row Score C"] <= 5:
        return 0
    else:
        return 1
```

```python
Output = df.apply(pin_update, axis=1)

df['Output'] = Output


df1 = df[["Anteroposterior 1.1",
          'Anteroposterior 1.2',
          'Vertical 1.1',
          'Vertical 1.2',
          'Transverse 1',
          'Crowding 1',
          'Spacing 1',
          'Tooth shape/size 1',
          'Pan 1',
          'Anteroposterior 2.1',
          'Anteroposterior 2.2',
          'Vertical 2.1',
          'Vertical 2.2',
          'Transverse 2',
          'Crowding 2',
          'Spacing 2',
          'Tooth shape/size 2',
           'Pan 2',
          'Output']]



df1 = df1.iloc[1:, :]

df1 = df1.dropna()

target_column = ["Output"]
numerical_column = df1.columns.drop(target_column)
output_rows = df1[target_column]
df1.drop(target_column,axis=1,inplace=True)

scaler = MinMaxScaler()
df1 = scaler.fit_transform(df1)

X_train, X_test, y_train, y_test = train_test_split(df1, output_rows, test_size=0.25, shuffle=True)
```

```python
acc = []
prediction = []
confusion_matrix = []

from sklearn.ensemble import RandomForestClassifier

for i in range(25):
    X_train, X_test, y_train, y_test = train_test_split(df1, output_rows, test_size=0.25, shuffle=True)
    
    clf = RandomForestClassifier()
    
    clf = clf.fit(X_train, y_train.values.ravel())
    
    pre = clf.predict(X_test)
    
    cm = metrics.confusion_matrix(y_test, pre)
    
    acc.append(clf.score(X_test, y_test))
    confusion_matrix.append(cm)
    prediction.append(pre)
    
print(f"Accuracy: {np.mean(np.array(acc))}")
print(f"Confusion matrix: \n {sum(confusion_matrix)}")

```

```python
xnew = np.array([[12, 0,0,0,8,2,0,1,4,
                  12,2,0,4,8,3,0,0,4]])

xnew = scaler.transform(xnew)
clf.predict_proba(xnew)
```

```python
acc = []
prediction = []
confusion_matrix = []

from sklearn.tree import DecisionTreeClassifier
for i in range(25):
    X_train, X_test, y_train, y_test = train_test_split(df1, output_rows, test_size=0.25, shuffle=True)
    
    clf = DecisionTreeClassifier()
    
    clf = clf.fit(X_train, y_train.values.ravel())
    
    pre = clf.predict(X_test)
    
    cm = metrics.confusion_matrix(y_test, pre)
    
    acc.append(clf.score(X_test, y_test))
    confusion_matrix.append(cm)
    prediction.append(pre)
    
print(f"Accuracy: {np.mean(np.array(acc))}")
print(f"Confusion matrix: \n {sum(confusion_matrix)}")
```

```python
xnew = np.array([[12, 0,0,0,8,2,0,1,4,
                  12,2,0,4,8,3,0,0,4]])

xnew = scaler.transform(xnew)

clf.predict_proba(xnew)
```

```python

acc = []
prediction = []
confusion_matrix = []

from sklearn.neighbors import KNeighborsClassifier

for i in range(25):
    X_train, X_test, y_train, y_test = train_test_split(df1, output_rows, test_size=0.25, shuffle=True)
    
    clf = KNeighborsClassifier()
    
    clf = clf.fit(X_train, y_train.values.ravel())
    
    pre = clf.predict(X_test)
    
    cm = metrics.confusion_matrix(y_test, pre)
    
    acc.append(clf.score(X_test, y_test))
    confusion_matrix.append(cm)
    prediction.append(pre)
    
print(f"Accuracy: {np.mean(np.array(acc))}")
print(f"Confusion matrix: \n {sum(confusion_matrix)}")
```
